```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


temperaturas = np.round(np.random.uniform(20, 35, 7),1)
diasSemana = np.array(['Segunda-feira', 'Terça-feira', 'Quarta-feira', 'Quinta-feira', 'Sexta-feira', 'Sábado', 'Domingo'])

data = {
    'Temperatura': temperaturas,
    'Dia da Semana': diasSemana
}

df = pd.DataFrame(data)
print(df)

soma = 0
for temperatura in temperaturas:
  soma = soma + temperatura

media = soma/7
print(f'a média das temperaturas na semana é: ', media, '°C')


x=df['Dia da Semana']
y=df['Temperatura']
plt.xlabel('Dias da Semana')
plt.ylabel('Temperaturas(em °C)')
plt.title('Variação da temperatura na semana')
plt.plot(x,y)
plt.xticks(rotation=15)
plt.show()
```
