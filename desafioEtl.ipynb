```python
import pandas as pd

df = pd.read_csv('/content/dados_nao_tratados.csv')
df
```

```python
df["ativo"] = df["ativo"].str.strip().map({'sim': True,'Não':False, 'Sim': True,'TRUE': True, 'TRUE': True, 'e': None,'FALSE': False,'nao': False,'Yes': True })
df
```

```python
df.replace('dez', 10, inplace=True)
df['nota'] = pd.to_numeric(df['nota'], errors='coerce')
df
```

```python
df['data_inscricao'] = pd.to_datetime(df['data_inscricao'], errors='coerce')
df
```

```python
df.replace('vinte e três', 23, inplace=True)
df["idade"] = pd.to_numeric(df["idade"], errors="coerce")
df
```
